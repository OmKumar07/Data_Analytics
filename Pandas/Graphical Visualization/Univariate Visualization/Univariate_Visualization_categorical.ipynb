```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head(2)
```

## BAR GRAPH

```python
sns.countplot(data=df , x=df["neighbourhood_group"])
```

```python
sns.countplot(data=df , x="room_type",color='tomato')
```

```python
sns.countplot(data=df, x="neighbourhood_group",hue="room_type")
```

## PIE CHART

```python
df["room_type"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

```python
df["neighbourhood_group"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

```python
df["neighbourhood_group"].value_counts().plot(kind = "bar")
```
