```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head(3)
```

## Histogram

```python
sns.histplot(df["price"])
```

```python
sns.histplot(data=df[df["price"]<500],x = 'price')
```

```python
sns.histplot(df["availability_365"])
```

```python
sns.distplot(df['price'])   ##histogram + probability distribution
```

```python
sns.distplot(df["availability_365"])
```

```python
sns.distplot(df["reviews_per_month"])
```

## Boxplot

```python
sns.boxplot(df[df["price"]<500]["price"],color='mediumvioletred')
```

```python
sns.boxplot(df["availability_365"],color='slategray')
```

```python
sns.violinplot(df["availability_365"],color='orchid') ##shows distribution as well as outliers
```
