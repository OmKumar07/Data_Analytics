```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.shape
```

```python
df.head()
```

```python
sns.barplot(x = 'class', y='age',data=df)
```

```python
sns.barplot(x = 'class', y='fare',data=df,estimator='sum')
```

```python
sns.barplot(x = 'class', y='fare',data=df,hue='sex')
```

```python
sns.barplot(x = 'class', y='fare',data=df,estimator='sum',hue='sex')
```

```python
sns.boxplot(x = 'class', y='age',data=df)
```

```python
sns.boxplot(x = 'class', y='age',hue='survived',data=df)
```

```python
sns.violinplot(x = 'class', y='age',data=df,color="goldenrod")
```
