## First Method Drop Rows With nan

```python
df.dropna()
```

## Replacing With Average Value (For Numerics)

### Imputers

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean') ##can use mod most freq etcc
```

### Slicing the Data frame using i loc

```python
df.iloc[:,1:3].values
```

```python
imputer.fit(df.iloc[:,1:3].values)
x = imputer.transform(df.iloc[:,1:3].values)
x
```

```python
df.iloc[:,1:3] = x
df
```

## Replacing Null Values For Non Numerical

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer.fit(df.iloc[:,:1].values)
df.iloc[:,:1] = imputer.transform(df.iloc[:,:1].values)
```

```python
df
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer.fit(df.iloc[:,3:].values)
df.iloc[:,3:] = imputer.transform(df.iloc[:,3:].values)
```

```python
df
```

### For Whole Data Frame

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer.fit(df)
df = imputer.transform(df)
```

```python
df
```
