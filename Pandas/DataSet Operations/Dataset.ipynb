## Finding No. Of Apps With Rating = 5

```python
c = 0
for i in df['Rating']:
    if(i == 5):
        c+=1
print(c)
```

## Average App Reviews

```python
r = 0
for i in df['Reviews']:
    r+=int(i)
print(int(r/len(df['Reviews'])))
```

## Number of apps in Art And Design Category

```python
c = 0
for i in df['Category']:
    if (i == 'ART_AND_DESIGN'):
        c+=1
print(c)
```

## Free And PAid Application Count

```python
f = 0
for i in df['Type']:
    if i == 'Free':
        f+=1
print(f,'Free Applictions')
print(len(df['Type'])-f,'Paid Applications')
```

## Total Number Of Apps In  Each Category

```python
categories = {}

for name in df['Category'].unique():
    ct = 0
    for i in df['Category']:
        if(i == name):
            ct += 1
    categories[name] = ct
    
print(categories)
```
