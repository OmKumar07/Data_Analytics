```python
import json
import time
import math

fd = open('Record.json','r')
js = fd.read()
fd.close()

record = json.loads(js)
sale = ''

print("--------------------MENU---------------------")
for key in record.keys():
    print(key, record[key]['Name'],'|', record[key]['Price'],'|', record[key]['Qn'])
print("---------------------------------------------")
print('')

ui_name = str(input("Enter your name    : "))
ui_mail = str(input("Enter Mail ID      : "))
ui_ph   = str(input("Enter Phone No     :  "))
ui_pr   = str(input("Enter product ID   : "))
ui_qn   = int(input("Enter Quantity     : "))
print('')

if(record[ui_pr]['Qn'] >= ui_qn):
    
    
    print("------------------ Bill ---------------------")
    print('                                         ')
    
    print("Name         : ", record[ui_pr]["Name"])
    print("Price (Rs)   : ", record[ui_pr]["Price"])
    print("Quantity     : ", ui_qn)
    print("---------------------------------------------")
    print("Billing      : ", ui_qn * record[ui_pr]["Price"], "Rs")
    discount = 0.1*ui_qn * record[ui_pr]["Price"]
    print("Discount     : ",discount)
    print("---------------------------------------------")
    print('Paid         : ',ui_qn * record[ui_pr]["Price"]-discount)
    print("---------------------------------------------")

    record[ui_pr]['Qn'] = record[ui_pr]['Qn'] - ui_qn
    sale = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(ui_qn)+","+str(record[ui_pr]["Price"])+","+str(ui_qn * record[ui_pr]["Price"]-discount)+","+time.ctime()+'\n'
    fd = open('sales2.txt','a')
    fd.write(sale)
    fd.close()

elif(record[ui_pr]['Qn'] == 0):
    print("Sorry We Are Out Of Stock")
else:
    print("Sorry We are not Having ebough Quantity")
    print('We are only having',record[ui_pr]['Qn'],'Quantity')
    ch = str(input('Press Y/N to Purcahse: '))
    if(ch == 'y' or ch == 'Y'):
        
        print("------------------ Bill ---------------------")
        print('                                         ')
        
        print("Name         : ", record[ui_pr]["Name"])
        print("Price (Rs)   : ", record[ui_pr]["Price"])
        print("Quantity     : ", record[ui_pr]['Qn'])
        print("---------------------------------------------")
        print("Billing      : ", ui_qn * record[ui_pr]["Price"], "Rs")
        discount = 0.1*ui_qn * record[ui_pr]["Price"]
        print("Discount     : ",discount)
        print("---------------------------------------------")
        print('Paid         : ',ui_qn * record[ui_pr]["Price"]-discount)
        print("---------------------------------------------")
        
        sale = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(record[ui_pr]['Qn'])+","+str(record[ui_pr]["Price"]-discount)+","+str(record[ui_pr]['Qn'] * record[ui_pr]["Price"])+","+time.ctime()+'\n'
        record[ui_pr]['Qn'] = 0
        fd = open('sales2.txt','a')
        fd.write(sale)
        fd.close()
    
js = json.dumps(record)

fd = open('Record.json','w')
fd.write(js)
fd.close()

print('')
print("---------------------------------------------")
print("  Thanks for your order, Inventory Updated!  ")
print("---------------------------------------------")
```
